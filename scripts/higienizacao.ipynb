```python
import pandas as pd
import json
import os
```

```python
x = []


# for file in os.listdir("./etis/"):
#   filepath = os.path.join("./etis/", file)
#   with open(filepath, 'r') as file:
#     data = json.load(file)
#     x.append(data)


filepath = '../etis/4868.json'


with open(filepath, 'r') as file:
     data = json.load(file)
     x.append(data)
```

```python
dados_normalizados = pd.json_normalize(x)
dados_normalizados.head()
```

```python
print(len((dados_normalizados['autarquias'][0])))
```

```python
orgao = "autarquias" if len((dados_normalizados['autarquias'][0])) > 0 else "secretarias"

#dados_normalizados['autarquias'][0][0]['codigo']
dados_normalizados['id_orgao'] = dados_normalizados[orgao][0][0]['codigo']
dados_normalizados['sigla_orgao'] = dados_normalizados[orgao][0][0]['sigla']
dados_normalizados.drop(columns=['autarquias','secretarias'], inplace=True)
dados_normalizados['tipo_orgao'] = orgao
dados_normalizados
```

```python
dados_normalizados.drop(columns=['tags', 'metas', 'eixos', 'periodicidade.codigo', 'periodicidade.descricao'], inplace=True)
```

```python
dados_normalizados = dados_normalizados.explode('valores')
```

```python
dados_normalizados['valores']
```

```python
dados_normalizados['valores.valor'] = dados_normalizados['valores'].apply(lambda x: x['valor'])
dados_normalizados['valores.codigo'] = dados_normalizados['valores'].apply(lambda x: x['codigo'])
dados_normalizados['valores.data'] = dados_normalizados['valores'].apply(lambda x: x['data'])
#dados_normalizados['valores.descricao'] = dados_normalizados['descricao'].apply(lambda x: x['descricao'])
```
